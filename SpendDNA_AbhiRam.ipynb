```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Data set for DADS June.csv")

print("Dataset loaded successfully!")
print(df.head())
```

```python
# Clean column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Clean Amount
df["amount"] = df["amount"].astype(str)
df["amount"] = df["amount"].str.replace("₹", "")
df["amount"] = df["amount"].str.replace(",", "")
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

# Clean Balance
df["balance"] = df["balance"].astype(str)
df["balance"] = df["balance"].str.replace("₹", "")
df["balance"] = df["balance"].str.replace(",", "")
df["balance"] = pd.to_numeric(df["balance"], errors="coerce")

print("Data cleaned successfully!")
print(df.head())
```

```python
# Convert Date

df["date"] = pd.to_datetime(df["date"], errors="coerce", dayfirst=True)

# Standardize Type

df["type"] = df["type"].replace({
    "DR": "Debit",
    "CR": "Credit"
})

# Remove duplicate rows

before = len(df)

df = df.drop_duplicates()

after = len(df)

print("Duplicates Removed:", before - after)
print("Remaining Rows:", after)

print(df.dtypes)
```

```python
print(df["description"].head(30))
```

```python
def extract_vendor(desc):

    desc = str(desc).upper()

    if "SWIGGY" in desc or "BUNDL" in desc:
        return "Swiggy"

    elif "ZOMATO" in desc:
        return "Zomato"

    elif "ZEPTO" in desc:
        return "Zepto"

    elif "INSTAMART" in desc:
        return "Instamart"

    elif "BLINKIT" in desc:
        return "Blinkit"

    elif "UBER" in desc:
        return "Uber"

    elif "OLA" in desc or "ANI" in desc:
        return "Ola"

    elif "RAPIDO" in desc or "ROPPEN" in desc:
        return "Rapido"

    elif "STARBUCKS" in desc:
        return "Starbucks"

    elif "THIRDWAVE" in desc or "TWC" in desc:
        return "Third Wave Coffee"

    elif "COFFEE DAY" in desc or "CCD" in desc:
        return "Cafe Coffee Day"

    elif "TRUFFLES" in desc or "RESTAURANT" in desc or "EMPIRE" in desc or "MEGHANA" in desc or "DINEOUT" in desc:
        return "Restaurant"

    elif "AMAZON" in desc or "AMZN" in desc:
        return "Amazon"

    elif "FLIPKART" in desc or "FKART" in desc:
        return "Flipkart"

    elif "MYNTRA" in desc:
        return "Myntra"

    elif "NYKAA" in desc:
        return "Nykaa"

    elif "GROWW" in desc:
        return "Groww"

    elif "ZERODHA" in desc:
        return "Zerodha"

    elif "GROFERS" in desc or "BIGBASKET" in desc or "DMART" in desc or "KIRANAKART" in desc or "AVENUE" in desc:
        return "Groceries"

    elif "BMTC" in desc:
        return "BMTC"

    elif "BESCOM" in desc or "BWSSB" in desc:
        return "Utilities"

    elif "INDIAN OIL" in desc or "HP PETROL" in desc:
        return "Fuel"

    elif "NETFLIX" in desc:
        return "Netflix"

    elif "HOTSTAR" in desc or "STAR INDIA" in desc:
        return "Hotstar"

    elif "SPOTIFY" in desc:
        return "Spotify"

    elif "AIRTEL" in desc or "JIO" in desc or "VI POSTPAID" in desc:
        return "Mobile Recharge"

    elif "BMS" in desc:
        return "BookMyShow"

    elif "AMAN" in desc:
        return "P2P Transfer"

    elif "SALARY" in desc:
        return "Salary"

    else:
        return "Others"
```

```python
df["vendor_clean"] = df["description"].apply(extract_vendor)

print(df["vendor_clean"].value_counts().head(15))
```

```python
category = {
    "Swiggy": "Food Delivery",
    "Zepto": "Quick Commerce",
    "Blinkit": "Quick Commerce",
    "Amazon": "E-Commerce",
    "Flipkart": "E-Commerce",
    "Uber": "Transport",
    "Ola": "Transport",
    "BMTC": "Transport",
    "Groww": "Investment",
    "BookMyShow": "Entertainment",
    "Starbucks": "Cafe",
    "Third Wave Coffee": "Cafe",
    "Truffles": "Restaurant",
    "Salary": "Salary",
    "P2P Transfer": "Personal Transfer",
    "BESCOM": "Utilities"
}

df["category"] = df["vendor_clean"].map(category)

df["category"] = df["category"].fillna("Others")

print(df[["vendor_clean", "category"]].head(20))
```

```python
# Total Credits
total_credit = df[df["type"] == "Credit"]["amount"].sum()

# Total Debits
total_debit = df[df["type"] == "Debit"]["amount"].sum()

# Net Savings
net = total_credit - total_debit

# Savings Rate
saving_rate = (net / total_credit) * 100

print("Total Credits :", total_credit)
print("Total Debits  :", total_debit)
print("Net Savings   :", net)
print("Savings Rate  :", round(saving_rate, 2), "%")

print("\nTop 5 Categories")
print(df.groupby("category")["amount"].sum().sort_values(ascending=False).head())

print("\nTop 5 Vendors")
print(df.groupby("vendor_clean")["amount"].sum().sort_values(ascending=False).head())

print("\nTotal Transactions :", len(df))
```
